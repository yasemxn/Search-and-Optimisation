## Algorithm Selection

1. **Dijk's Optimization:** Dijkstra’s algorithm is an algorithm for finding single-source shortest path having non-negative edge weight in the graphs, that is, it's used to find the shortest distance between two vertices or edges on a graph (van Dijk et al., 2013).
    - Mark the initial point (source node) as 0 while setting other points as infinity.
    - From the source node, identify the neighbours, then assign the values of their transmission rate to replace the infinity earlier set only for those neighbours.
    - The next node will be the node with the least transmission from the source, for this node, the neighbours will be identified and their values will be the cumulation of rates from the source through the parent node and from its parent node to itself. The rates are recorded. This is repeated for other neighbouring nodes from the source. If a node has been passed through before, if the value of the new route is greater than the older value, it is discarded, however, if it is less, it will be the new transmission for the route.
    - Steps 2 and 3 are repeated for other nodes until the source nodes gets to the destination node which is either of the base station.
<br>
<br>
2. **Genetic Algorithm (GA):** This uses a combination of selection, crossover, and mutation to evolve the population over generations, with the goal of finding a solution with both a short path and a high transmission rate (Katoch S. et al., 2020).
    
   - **Individual Representation:**
       - Define a class `Individual` representing an individual in the population and calculate fitness i.e Minimum Transmission rate.

   - **Population Initialization:**
       - Each individual's path is randomly generated until it reaches one of the base stations
       - Discard paths that do not reach any of the target nodes.

   - **Genetic Algorithm Loop:**
       - Execute the genetic algorithm loop for a specified number of generations.

   - **Crossover Operation:**
       - Perform crossover by taking half of the path from the fittest individual and assigning it to other individuals in the population.

   - **Mutation Operation:**
       - Apply mutation by randomly changing one node in the path of each individual.

   - **Sort Population:**
       - Sort the population again based on fitness values after crossover and mutation.

   - **Select Best Solution:**
       - Identify the best path and maximum transmission rate from the fittest individual in the final population.

   - **Run the Genetic Algorithm:**
       - Execute the genetic algorithm by calling the `genetic_algorithm` function with specified parameters for population size and number of generations.

## Data Processing
**1. Distance and Transmission rate Measurement**

```python
def distance(p1, p2):
    distance = np.sqrt((p1[0]-p2[0])**2 +(p1[1]-p2[1])**2 )
    if distance < 500:
        return 10
    elif distance <1000:
        return 8
    elif distance <2000:
        return 6
    elif distance < 3000:
        return 4
    elif distance < 4000:
        return 2
    elif distance < 6000:
        return 1
    else:
        return 0
```

The function `distance` takes in two arguments point 1 and Point 2 where each one represents a point in this format `(x, y)`.
The output of the function is the transmission rate between the two points

**2. Network Topology**

```python
def find_graph(df):
    graph = {}
    df = df.dropna(subset=['x_axis', 'y_axis'])

    for i, row_i in df.iterrows():
        vals = {}
        for j, row_j in df.iterrows():
            if i != j:
                t_rate = distance((row_i['x_axis'], row_i['y_axis']),
                                  (row_j['x_axis'], row_j['y_axis']))
                if t_rate != 0:
                    vals[j + 1] = t_rate
        graph[i + 1] = vals

    return graph
```

The function `find_graph` takes in an argument `df` which represents the data in the dataset.
Then for each car in the data, it finds the cars which they are linked to i.e cars with which their transmission rate is not zero.
The final output is a graph which maps of each car to adjascent cars that it can transmit to.

## Dijks Algorithm

```python
import heapq
import json

def dijks_algo(transmission_rates, start, end):
    max_rates = {node: float('-inf') for node in transmission_rates}
    max_rates[start] = float('inf')
    heap = [(-float('inf'), start, [])]

    while heap:
        current_rate, current_node, current_path = heapq.heappop(heap)

        if current_node == end:
            return current_path + [end], -max_rates[end]

        for neighbor, rate in transmission_rates[current_node].items():
            new_rate = min(-current_rate, rate)
            if new_rate > max_rates[neighbor]:
                max_rates[neighbor] = new_rate
                heapq.heappush(heap, (-new_rate, neighbor, current_path + [current_node]))

    return None, None

# Run the modified Dijkstra's algorithm for each base station and find the maximum maximum rate
#np.random.seed(42)
max_max_rate = float('-inf')
best_path = None

# Define the graph
graph = find_graph(df)

# Dictionary to store paths
all_paths = {}

# Iterate over all cars
for car_id in df['car_id']:
    start = car_id
    max_max_rate = float('-inf')
    best_path = None

    for base_station in {101, 102}:
        path, max_rate = dijks_algo(graph, start, base_station)
        if path is not None and -max_rate > max_max_rate:
            max_max_rate = -max_rate
            best_path = path

    if best_path is not None:
        path_info = route(graph, best_path)
        all_paths[f"Car-{car_id}"] = path_info

# Save the paths to a JSON file
with open('paths.json', 'w') as json_file:
    json.dump(all_paths, json_file, indent=2)

print("Paths saved to 'paths.json'")
```

The Algorithm was implemented using two functions `dijks_route` function which gets the routes starting from the source and the `spt` function which finds the shortest path in that route. This is iteratively carried out until the destination node is reached.

## Breadth-first search Algorithm

```python
from collections import deque
import json

def bfs_algo(transmission_rates, start):
    visited = set()
    queue = deque([(start, [start])])

    while queue:
        current_node, current_path = queue.popleft()

        if current_node not in visited:
            visited.add(current_node)

            # Process the node (you can perform additional actions here if needed)

            for neighbor in transmission_rates[current_node]:
                if neighbor not in visited:
                    queue.append((neighbor, current_path + [neighbor]))

    return current_path  # Return the path obtained during BFS

# Run BFS for each car and save paths to a JSON file
all_paths_bfs = {}


for car_id in df['car_id']:
    start_node = car_id
    try:
        bfs_result = bfs_algo(graph, start_node)
    except:
        # Define the graph
        graph = find_graph(df)

    if bfs_result is not None:
        path_info = route(graph, bfs_result)
        all_paths_bfs[f"Car-{car_id}"] = path_info

# Save the paths to a JSON file
with open('paths_bfs.json', 'w') as json_file:
    json.dump(all_paths_bfs, json_file, indent=2)

print("Paths saved to 'paths_bfs.json'")

```

## Genetic Algorithm

```python
class Individual:
    def __init__(self, path):
        self.path = path
        self.fitness = self.calculate_fitness()

    def __repr__(self):
        return f"path: {self.path}, fitness: {self.fitness}"

    def calculate_fitness(self):
        min_rate = float('inf')
        for i in range(len(self.path) - 1):
            source, target = self.path[i], self.path[i + 1]
            if target in graph.get(source, {}):
                rate = graph[source][target]
                min_rate = min(min_rate, rate)
        return min_rate

def initialize_population(population_size):
    population = []
    for _ in range(population_size):
        current_node = start
        path = [current_node]
        while current_node not in {101, 102}:
            next_nodes = [node for node in graph[current_node].keys() if node not in path]
            if not next_nodes:
                break
            current_node = random.choice(next_nodes)
            path.append(current_node)
        if (101 not in path) and (102 not in path):
            continue
        individual = Individual(path)
        population.append(individual)
    return population

def genetic_algorithm(start, population_size, generations):
    population = initialize_population(population_size)

    #The new population size is the number of population received from
    population_size = len(population)

    for generation in range(generations):
        population.sort(key=lambda x: x.fitness, reverse=True)
        fittest_individual = population[0]

        # Crossover (take half from the fittest individual)
        crossover_point = len(fittest_individual.path) // 2

        for i in range(1, population_size):
            population[i].path[:crossover_point] = fittest_individual.path[:crossover_point]

        # Mutation (randomly change one node in the path)
        for i in range(1, population_size):
            try:
                mutation_point = random.randint(1, len(population[i].path) - 2)
            except:
                mutation_point = 1
            current_node = population[i].path[mutation_point - 1]
            next_nodes = [node for node in graph[current_node].keys() if node not in population[i].path]
            if not next_nodes:
                continue
            new_node = random.choice(next_nodes)
            population[i].path[mutation_point] = new_node

    population.sort(key=lambda x: x.fitness, reverse=True)
    best_path = population[0].path
    max_transmission_rate = population[0].fitness

    return best_path, max_transmission_rate
graph = find_graph(df)

# Run the genetic algorithm
solutions = []
for _ in range(50):
    try:
        best_path, max_transmission_rate = genetic_algorithm(start,population_size=50, generations=100)
    except:
        start = np.random.randint(100)
        while start <= 0:
            start = np.random.randint(100)
        best_path, max_transmission_rate = genetic_algorithm(start,population_size=50, generations=100)
    solutions.append((best_path, max_transmission_rate))

try:
    shortest_path_solution = min(solutions, key=lambda x: len(x[0]))
except:
    shortest_path_solution = [None, None]

print(route(graph, shortest_path_solution[0]))
viz_route(graph, shortest_path_solution[0])
```

By utilizing vehicles as intermediaries, signals can be relayed and connected to strategically positioned base stations, potentially revolutionizing vehicle network communication.

This positioning enables us to provide reliable Internet access to vehicles across the entire motorway, ensuring seamless connectivity throughout the journey.
